```python
#Oscar Alejandro Reyna Monreal A01570385

import agentpy as ap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import IPython
from Modelos.Warehouse import Warehouse

parameters = {
   'size': 20,
   'robots': 5,
   'boxes': 40,
}

warehouse = Warehouse(parameters)

def animation_plot(model, ax):
   attr_grid = model.grid.attr_grid("condition")
   color_dict = {0:'#d62c2c', 1:'#666666', 2:'#000000', 3:'#808080', 4:'#666666', 5:'#595959', 6: '#404040', 7:'#262626', None:'#d5e5d5'}
   ap.gridplot(attr_grid, ax=ax, color_dict=color_dict, convert=True)
   ax.set_title(f"Warehouse Simulation\n"
               f"Time-step: {warehouse.t}")
   
fig, ax = plt.subplots()
animation = ap.animate(warehouse, fig, ax, animation_plot)
IPython.display.HTML(animation.to_jshtml(fps=15))
```
